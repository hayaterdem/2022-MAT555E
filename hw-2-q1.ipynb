```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.datasets import copper
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

prices = copper.data.load_pandas()
prices['data']

```

```python
# Assume that alfa is 0,05
data = prices['data']

model = ols('COPPERPRICE ~ WORLDCONSUMPTION + INCOMEINDEX + ALUMPRICE + INVENTORYINDEX + TIME', data).fit()
model.summary()
```

null test: If t* > t table value
H0: Coef is significant
H1: Coef is not significant

We can say INCOMEINDEX is not significant for this model. (3.555 < 54.635)
