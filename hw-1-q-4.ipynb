```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline

N = int(input("N="))
while(True):
    A = np.random.rand(N,1)
    B = np.square(A)
    C = sum(B)
    if(C == 1):
        break    
print(A)
print(C)
plt.plot(A,C)
plt.grid()
plt.title('Variables')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```
